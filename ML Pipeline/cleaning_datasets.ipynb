```python
def clean_dataset(filepath):
    df = pd.read_csv(filepath)
    tot_cols_to_remove = ['Unnamed: 0',
 'STROKE',
 'HETRA',
 'PARKISM',
 'RLS',
 'NARCLPSY',
 'DEPRS',
 'EPILEPSY',
 'BRNINFM',
 'CNSOTH',
 'SDMTOTAL',
 'NP3TOT',
 'NP2PTOT',
 'NP1PTOT',
 'NP1RTOT',
 'CNTRLDSM',
 'TMDISMED',
 'PDTRTMNT',
 'PDMEDYN',
 'DBSYN',
 'NUPSOURC',
 'PTCGBOTH',
 'RVWNPSY',
 'HVLTVRSN',
 'PTINBOTH',
 'SDMTVRSN']
    df_cols = df.columns.tolist()
    for col_to_remove in tot_cols_to_remove:
        if col_to_remove in df_cols:
            df_cols.remove(col_to_remove)
            print('successfully removed')

    new_filepath = filepath.split(".")[0] + "_2.csv"
    df = df[df_cols]
    df.to_csv(new_filepath, index=False)
```

```python
ls
```

```python
clean_dataset('only_motor.csv')
```

```python
clean_dataset('only_non_motor.csv')
```

```python
clean_dataset('only_self_report.csv')
```
