```python
def find_feature(f_name, f_type):
    # f -> feature
    current_dir = os.getcwd()
    target_dir = os.path.join(current_dir, f_type)
    item_dirs = os.listdir(target_dir)
    file_dirs = [file_dir for file_dir in item_dirs if file_dir.endswith('.csv')]
    for file_dir in file_dirs:
        file_path = os.path.join(target_dir, file_dir)
        df = pd.read_csv(file_path)
        cols = df.columns.tolist()
        if f_name in cols:
            return file_dir
```

```python
for row in schema_df.itertuples():
    
    if row.TYPE == 'Motor':
        file_name = find_feature(row.ITM_NAME, row.TYPE)
        file_names_temp.append(file_name)
    elif row.TYPE == 'Non Motor':
        file_name = find_feature(row.ITM_NAME, row.TYPE)
        file_names_temp.append(file_name)
    else:
        file_name = find_feature(row.ITM_NAME, 'Motor')
        if file_name == None:
            file_name = find_feature(row.ITM_NAME, 'Non Motor')
        file_names_temp.append(file_name)

print('Done!')
file_names = file_names_temp
file_names_temp = []
```
